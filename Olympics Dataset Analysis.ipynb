```python
## load dataset
athletes = pd.read_csv("athlete_events.csv")
regions = pd.read_csv("noc_regions.csv")
```

```python
athletes.head()
```

```python
regions.head()
```

```python
# join the datframes
athletes_df = athletes.merge(regions, how = 'left', on ='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
#columns names consistence
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace=True)
athletes_df.head()
```

```python
## top countries participating
top_10_counties = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)
top_10_counties
```

```python
## plot for the top 10 countries
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# top 10 countries participating
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)

# plot for the top 10 countries
plt.figure(figsize=(12, 6))
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2')
plt.xticks(rotation=20)  # Optional: Rotate x-axis labels for better readability
plt.show()
```

```python
# Age Distributes of the Participants 

plt.figure(figsize=(12,6))
plt.title("Age distributions of the athletes")
plt.xlabel("Age")
plt.ylabel("Number of participants")
plt.hist(athletes_df.Age , bins = np.arange(10,80,2), color='yellow' ,edgecolor ='white');
```

```python
winter_sports= athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
## summer olympics sports
summer_sports= athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
## find the male and female participants
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python

#pie plot for male and female athletes
plt.figure(figsize=(12, 6))
plt.title("Gender Distribution")
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=150, shadow=True)
plt.show()
```

```python
## Total number of female athletes in each olympics
female_participatings = athletes_df[(athletes_df['Sex'] == 'F') & (athletes_df['Season'] == 'Summer')][['Sex', 'Year']]
female_participatings = female_participatings.groupby('Year').count().reset_index()
female_participatings.head()
```

```python
women_Olympics = athletes_df[(athletes_df.Sex =='F')&(athletes_df.Season == 'Summer')]
women_Olympics
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data =women_Olympics, palette='Spectral')
plt.title('women Participation')
```

```python
part = women_Olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('plot of female athletes over time')           
```

```python
# Assuming 'df' is your original dataset and 'Medal' column has values like 'Gold', 'Silver', 'Bronze'
Total_Gold_Medals = athletes_df[athletes_df['Medal'] == 'Gold']
Total_Gold_Medals
```

```python

Total_Gold_Medals = Total_Gold_Medals.Region.value_counts().reset_index(name='Medal').head(6)
Total_Gold_Medals
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

g = sns.catplot(
    x="Region", 
    y="Medal", 
    data=Total_Gold_Medals, 
    height=5, 
    kind="bar", 
    palette="rocket"
)
g.despine(left=True)
g.set_xlabels("Top 10 Countries")
g.set_ylabels("Number of Medals")
plt.title("Gold Medals per Country")
#plt.xticks(rotation=45)  # Optional: improves readability
plt.show()


```
